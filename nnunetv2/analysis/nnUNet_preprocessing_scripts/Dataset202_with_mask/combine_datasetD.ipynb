```python
import os
import shutil
from pathlib import Path
```

```python
data_root = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Train"
data_root_D = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Train_D"

def combine_datasetD(data_root, data_root_D):
    data_root = Path(data_root)
    data_root_D = Path(data_root_D)
    for patient_dir in data_root_D.glob("Task*/*/*"):
        relative_path = patient_dir.relative_to(data_root_D)
        target_dir = data_root / relative_path
        if patient_dir.name == "overviews":
            target_dir = target_dir.with_name(target_dir.name + '_D')
        print(patient_dir, "=>", target_dir)
        if not target_dir.exists():
            shutil.copytree(patient_dir, target_dir)
        else:
            print(f"Patient {relative_path} already exists in the target directory.")

combine_datasetD(data_root, data_root_D)
```

```python
data_root = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Val_Input"
data_root_D = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Val_Input_D"

combine_datasetD(data_root, data_root_D)
```

```python
data_root = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task2_Val_Input"
data_root_D = "/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task2_Val_Input_D"

combine_datasetD(data_root, data_root_D)
```

# CHECK NUMBER OF PATIENTS

```python
# Compute the number of patients in each folder
def count_patients(data_root):
    data_root = Path(data_root)
    patient_count = 0
    for patient_dir in data_root.glob("Task*/*/*"):
        if patient_dir.is_dir() and not "overviews" in patient_dir.name:
            patient_count += 1
    print(f"Number of patients in {data_root.name}: {patient_count}")
    return patient_count

# should be 221 + 182+ 175 = 578
count_patients("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Train") 
# should be 34 + 28 + 27 = 89
count_patients("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task1_Val_Input")
# should be 325 + 321 + 309 = 955
count_patients("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task2_Train")
# should be 50 + 50 + 48 = 148
count_patients("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_Task2_Val_Input")
```

# Move datasetD to backup

```python
# move the D dataset to another folder
data_path = Path("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/synthRAD2025_*_D")
datasetD_path = Path("/datasets/work/hb-synthrad2023/source/synthrad2025_data_v2/backup_datasetD")
datasetD_path.mkdir(parents=True, exist_ok=True)
for path in data_path.glob("*"):
    if path.is_dir():
        target_path = datasetD_path / path.name
        if not target_path.exists():
            shutil.move(path, target_path)
        else:
            print(f"Directory {target_path} already exists. Skipping move.")
```
